```python
print('esta es una calculadora') 
print(' elije el tipo de operacion de menu')
print('1.sumar')
print('2. restar')
print('3. multiplicar')
print('4. dividir')
res=input('¿que quieres hacer?')
num1=float(input('escribe el primero numero'))
num2=float(input('escribe el segundo numero'))
if res=='sumar':
  print('{} + {} = {}'.format(num1,num2,num1+num2))
elif res=='restar':
  print('{} - {} = {}'.format(num1,num2,num1-num2))
elif res=='multiplicar':
  print('{} * {} = {}'.format(num1,num2,num1*num2))
elif res=='dividir':
  print('{} / {} = {}'.format(num1,num2,num1/num2))
else:
 print('comando no valido')
```
